# Importing Libraries

```python
import matplotlib.pyplot as plt
import json
from scipy.spatial import ConvexHull, convex_hull_plot_2d
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

# Constants

```python
NUM_ZONES = 5
NUM_TIMESLOTS = 1440
```

# Clustering

```python
# returns dbscan clusters
def dbscan(X, eps, min_samples):
    db = DBSCAN(eps=eps, min_samples=min_samples)
    cluster = db.fit(X)
    return cluster

# returns kmeans clusters
def kmeans(X, number_clusters):
    kmeans = KMeans(n_clusters = number_clusters)
    cluster = kmeans.fit(X)
    return cluster
```

# Cluster Boundary Acquisition (DBSCAN)

```python
def dbscan_get_cluster(dataframe, eps, min_samples):
    list_cluster = []
        
    count = 0
    for zone in range(NUM_ZONES):
        mod_dataframe = dataframe[['Occupant\'s Zone','Zone Arrival Time (Minute)', 'Stay Duration (Minute)']].values
        features = mod_dataframe[mod_dataframe[:,0] == zone][:,1:]
        
        cluster_model = dbscan(features, eps, min_samples)
        labels = cluster_model.labels_
                
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        for cluster in range(n_clusters):
            points = []
            for k in range(len(labels)):
                if labels[k] == cluster:
                    points.append(features[k])

            if len(points) >= 3:
                try:
                    vertices = convex_hull(zone, points)       
                except:
                    pass
            list_cluster.append({"zone_id": zone, "cluster_id": cluster, "points": vertices})
            
    return list_cluster
```

# Cluster Boundary Acquisition (KMeans)

```python
def kmeans_get_cluster(dataframe, number_clusters):
    list_cluster = []
        
    count = 0
    for zone in range(NUM_ZONES):
        mod_dataframe = dataframe[['Occupant\'s Zone','Zone Arrival Time (Minute)', 'Stay Duration (Minute)']].values
        features = mod_dataframe[mod_dataframe[:,0] == zone][:,1:]
        
        cluster_model = kmeans(features, number_clusters)
        labels = cluster_model.labels_
                
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        for cluster in range(n_clusters):
            points = []
            vertices = []
            
            for k in range(len(labels)):
                if labels[k] == cluster:
                    points.append(features[k])
            
            if len(points) >= 3:
                try:
                    vertices = convex_hull(zone, points)       
                except:
                    pass
            list_cluster.append({"zone_id": zone, "cluster_id": cluster, "points": vertices})
            
    return list_cluster
```

# Plotting Function

```python
def plotting(list_cluster, z_lim):
    #ax = plt.figure().add_subplot(projection='3d')
    from matplotlib.collections import PolyCollection
    fig= plt.figure(figsize=(8,5.5))
    ax = plt.axes(projection='3d')

    legendlist=['Outside', 'Bedroom', 'Livingroom', 'Kitchen', 'Bathroom']

    colors = ['#1B77BD', '#F87F00', '#369B00', '#D12D21', '#8E68C2']

    color_index = 0
    for i in range(len(list_cluster)):
        try:
            plt.plot(*zip(*list_cluster[i]["points"]), colors[color_index], alpha = 0.8)
        except:
            pass
        if i < len(list_cluster) - 1:
            if list_cluster[i]["zone_id"] != list_cluster[i + 1]["zone_id"]: 
                color_index += 1

    plt.legend(legendlist,loc='best',fontsize="10")

    plt.gca().get_legend().legendHandles[0].set_color(colors[0])
    plt.gca().get_legend().legendHandles[1].set_color(colors[1])
    plt.gca().get_legend().legendHandles[2].set_color(colors[2])
    plt.gca().get_legend().legendHandles[3].set_color(colors[3])
    plt.gca().get_legend().legendHandles[4].set_color(colors[4])

    plt.yticks(fontsize="15")
    plt.xticks(fontsize="15")
    ax.zaxis.set_tick_params(labelsize=15)

    ax.set_xlabel("Time of the Day (minute)",fontsize="15")
    ax.set_ylabel("Zone",fontsize="15")
    ax.set_zlabel("Duration of Stay (minute)",fontsize="15")


    ytick_labels = ['Z-0', 'Z-1', 'Z-2', 'Z-3', 'Z-4']
    plt.yticks(range(len(ytick_labels)), ytick_labels)
    #ax.set_yticks([0,1,2,3,4])
    plt.yticks(range(len(ytick_labels)), ytick_labels)
    
    ax.set_zlim(0,z_lim)
    ax.azim = 20
    ax.dist = 7
    ax.elev = 20
```

# DBSCAN Plotting Cluster for House-A and Occupant-1

```python
cleaned_dataframe_house_A_occ_1 = pd.read_csv("../../data/cleaned/Cleaned-Dataframe_House-A_Occupant-1.csv")
cleaned_dataframe_house_A_occ_1
list_cluster = dbscan_get_cluster(cleaned_dataframe_house_A_occ_1, 30, 3)

plotting(list_cluster, 100)
plt.savefig("../../plots/DBSCAN_Clusters-House_A_Occ-1" + '.pdf', bbox_inches='tight')
```
