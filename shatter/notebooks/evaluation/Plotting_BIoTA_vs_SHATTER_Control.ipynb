# Plotting Function

```python
def plotting(biota_data, shatter_data, y_max, location):
    legendlist=['ASHRAE Control Cost','SHATTER Control Cost' ]
    markers=['v','o','s']
    fig= plt.figure(figsize=(8,5))
    plt.grid()
    plt.plot(biota_data, marker=markers[0], linewidth=3, markersize=12)
    plt.plot(shatter_data, marker=markers[1], linewidth=3, markersize=12)
    plt.legend(legendlist,loc='best',fontsize="20")
    plt.yticks(fontsize="20")
    plt.xticks(fontsize="20")
    plt.ylim(0, y_max)
    plt.xlabel("Day of the Month",fontsize="24")
    plt.ylabel("Control Cost ($)",fontsize="24")
    plt.savefig(location + '.pdf', bbox_inches='tight')
```

# House-A Control Comparison Figure

```python
y_max = 12
location = "../../plots/House-A_ASHRAE-vs-SHATTER-Control-Cost"
plotting(day_wise_biota_benign_cost_house_A, day_wise_shatter_cost_house_A, y_max, location)
```

# House-B Control Comparison Figure

```python
y_max = 12
location = "../../plots/House-B_ASHRAE-vs-SHATTER-Control-Cost"
plotting(day_wise_biota_benign_cost_house_B, day_wise_shatter_cost_house_B, y_max, location)
```
