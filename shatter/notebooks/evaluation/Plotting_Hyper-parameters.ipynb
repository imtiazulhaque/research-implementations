# Importing Hyperparameter Learning data

```python
dbscan_house_A_occ_1 = pd.read_csv("../../data/shatter/DBSCAN_HyperParameter_House-A_Occ-1.csv")
dbscan_house_A_occ_2 = pd.read_csv("../../data/shatter/DBSCAN_HyperParameter_House-A_Occ-2.csv")
dbscan_house_B_occ_1 = pd.read_csv("../../data/shatter/DBSCAN_HyperParameter_House-B_Occ-1.csv")
dbscan_house_B_occ_2 = pd.read_csv("../../data/shatter/DBSCAN_HyperParameter_House-B_Occ-2.csv")
kmeans_house_A_occ_1 = pd.read_csv("../../data/shatter/KMeans_HyperParameter_House-A_Occ-1.csv")
kmeans_house_A_occ_2 = pd.read_csv("../../data/shatter/KMeans_HyperParameter_House-A_Occ-2.csv")
kmeans_house_B_occ_1 = pd.read_csv("../../data/shatter/KMeans_HyperParameter_House-B_Occ-1.csv")
kmeans_house_B_occ_2 = pd.read_csv("../../data/shatter/KMeans_HyperParameter_House-B_Occ-2.csv")
```

# Plotting DBSCAN Hyperparameters

```python
def plotting_dbscan(dataframe, y_max, location):
    legendlist=["Davies Bouldin Score", "Silhouette Score", "Calinski-Harabasz Index"]
    markers=['v','o','s']
    fig= plt.figure(figsize=(8,5))
    plt.style.use('default')
    plt.grid()
    plt.plot(dataframe["Silhouette Score"], marker=markers[0], linewidth=3, markersize=12)
    plt.plot(dataframe["Davies Bouldin Score"], marker=markers[1], linewidth=3, markersize=12)
    plt.plot(dataframe["Calinski-Harabasz Index"]/10, marker=markers[2], linewidth=3, markersize=12)
    plt.legend(legendlist,loc='upper right',fontsize="20")
    plt.yticks(fontsize="20")
    plt.xticks(fontsize="20")
    plt.ylim(0, y_max)
    plt.xlabel("Hyperparameter (minPts)",fontsize="24")
    plt.ylabel("Score",fontsize="24")
    plt.savefig(location + '.pdf', bbox_inches='tight')
```

# Plotting KMeans Hyperparameters

```python
def plotting_kmeans(dataframe, y_max, location):
    legendlist=["Davies Bouldin Score", "Silhouette Score", "Calinski-Harabasz Index"]
    markers=['v','o','s']
    fig= plt.figure(figsize=(8,5))
    plt.style.use('default')
    plt.grid()
    plt.plot(dataframe["Silhouette Score"], marker=markers[0], linewidth=3, markersize=12)
    plt.plot(dataframe["Davies Bouldin Score"], marker=markers[1], linewidth=3, markersize=12)
    plt.plot(dataframe["Calinski-Harabasz Index"]/1000, marker=markers[2], linewidth=3, markersize=12)
    plt.legend(legendlist,loc='upper right',fontsize="20")
    plt.yticks(fontsize="20")
    plt.xticks(fontsize="20")
    plt.ylim(0, y_max)
    plt.xlabel("Hyperparameter (k)",fontsize="24")
    plt.ylabel("Score",fontsize="24")
    plt.savefig(location + '.pdf', bbox_inches='tight')
```

# Plotting House-A Occupant-1 DBSCAN Hyperparameters

```python
plotting_dbscan(dbscan_house_A_occ_1.iloc[0:len(dbscan_house_A_occ_1):2,:], 27, '../../plots/Hyperparameter_DBSCAN_House_A_Occ_1')
```
